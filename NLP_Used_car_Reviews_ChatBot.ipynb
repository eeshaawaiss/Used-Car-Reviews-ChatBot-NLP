```python
import pandas as pd

df=pd.read_csv("/content/Pak Suzuki Car Reviews.csv")
df.head()
#importing the downloaded dataset
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head()
#removing "unnamed" coloumn
```

```python
import datetime as dt
df['Date']=pd.to_datetime(df['Date']) #converting date to datetime format (YYYY-MM-DD)
df['Year'] = df['Date'].dt.year # extracting year from date and adding it to new coloumn "Year"
df.head()
```

```python
df["Review"]=df["Review"].str.replace("\n", "") #replacing new line characters ("/n" in reviews coloumn with empty string)
df.head()
```

```python
rating_to_words = {
    1: 'Poor',
    2: 'Fair',
    3: 'Average',
    4: 'Good',
    5: 'Excellent'
} #making a dictionary to map numeric ratings to words

df[["Overall_rating", "Style", "Comfort", "Fuel_Economy", "Performance", "Value_for_Money" ]] = df[["Overall_rating", "Style", "Comfort", "Fuel_Economy", "Performance", "Value_for_Money"]].replace(rating_to_words)
#mapping word ratings to numeric in dataset
df.head()
```

```python
df['Review'] = df['Review'].str.replace("[?.,$]", " ") #replace punctuation characters with empty sttring in review coloumn
df.head()
```

```python
from langchain.document_loaders import CSVLoader #to load .csv file
```

```python
loader = CSVLoader(file_path="/content/Pak Suzuki Car Reviews.csv") #path to csv file
reviews = loader.load_and_split() #load and split file in individual documents
reviews[0] #first review document
```

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings #load huggingface model

model_name = "google/flan-t5-base" #use T5 model
model_kwargs = {"device": "cuda"} #use GPU
encode_kwargs = {"normalize_embeddings": True}
embeddings = HuggingFaceBgeEmbeddings(
    model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs
)
```

```python
from langchain.vectorstores import Chroma
#create vector store db from chroma lib
db = Chroma.from_documents(reviews,
                           embedding=embeddings,
                           persist_directory='./llama-db')
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, GPTQConfig
#load pre trained gpt
tokenizer = AutoTokenizer.from_pretrained("TheBloke/Llama-2-7b-Chat-GPTQ")
tokenizer.pad_token = tokenizer.eos_token
quantization_config_loading = GPTQConfig(bits=4, disable_exllama=True, tokenizer=tokenizer)
model = AutoModelForCausalLM.from_pretrained("TheBloke/Llama-2-7b-Chat-GPTQ",
                                              quantization_config=quantization_config_loading,
                                              device_map="auto")
```

```python
from transformers import pipeline
import torch
#implement text generating pipeline
pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 256,
                do_sample=True,
                top_k=1,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id,
                repetition_penalty=1.2
                )
```

```python
from langchain.llms import HuggingFacePipeline
#implementing llm from prev pipeline
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
print(llm(prompt="What are the reviews regarding Suzuki Alto in year 2015?")) #generating output result
```

```python
from langchain.chains import RetrievalQA
#qna using retrieval based method
rag = RetrievalQA.from_chain_type(
        llm=llm, chain_type='stuff',
        retriever=db.as_retriever()
        )
print(rag("What are the reviews regarding Suzuki Alto in year 2015?")["result"])
```

```python
question = "how is alto 2022?"
searchDocs = db.similarity_search(question)
print(searchDocs[0].page_content)
#searching similar contents relating to the qs and printing the first relevant result
```

```python
question_answer = pipeline(
    "question-answering",
    model=model_name,
    tokenizer=tokenizer,
    return_tensors='pt'
)#qna from huggingface model
```

```python
llm = HuggingFacePipeline(
    pipeline=question_answer,
    model_kwargs={"temperature": 0, "max_length": 256},
)
```

```python
retriever = db.as_retriever(search_kwargs={"k": 4})
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="refine", retriever=retriever, return_source_documents=False)
```

```python
question = "What are the reviews regarding Suzuki Alto in year 2015?"
answer = qa(question)["result"]
print(answer)
```
